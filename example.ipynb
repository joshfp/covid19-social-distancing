## Calibration

```python
import PIL.Image as Image
import PIL.ImageDraw as ImageDraw
import numpy as np

from distancing_detector import Perspective2Plane
import settings
```

### Calibration image

```python
img = Image.open('example_data/image.jpg')
img
```

### Visualize perspective & distance points

```python
img1 = img.copy()
draw = ImageDraw.Draw(img1, 'RGBA')
draw.polygon(settings.camera_calibration['perspective_pts'], fill=(255,0,0,70))
draw.line(settings.camera_calibration['distance_pts'], fill='green', width=10)
img1
```

### Perspective to plane

```python
p2p = Perspective2Plane(**settings.camera_calibration)
```

```python
Image.fromarray(p2p.transform_image(np.array(img1)))
```

We can see that camera is propertly calibrated
